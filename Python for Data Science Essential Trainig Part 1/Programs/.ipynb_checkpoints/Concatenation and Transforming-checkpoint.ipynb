```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.arange(0,36).reshape(6,6))
df
```

```python
df1=pd.DataFrame(np.arange(15).reshape(5,3))
df1
```

## Concatenation

```python
# axis=1 for cancatenate by adding columns
pd.concat([df,df1], axis=1)
```

```python
# concat by rows
pd.concat([df,df1])
```

## Transformation of Data

#### Droping Data

```python
# droping rows
df.drop([0,2])
```

```python
# droping columns
df.drop([0,2], axis=1)
```

#### Adding Data

```python
ss=pd.Series(np.arange(6))
ss.name = "Added Variable"   #used to give name to series object
ss
```

```python
# this will add series object to our DataFrame
v_add = pd.DataFrame.join(df,ss)
v_add
```

#### Sorting

```python
df_sorted = df.sort_values(by=(5),ascending=False)
df_sorted
```

```python
df_sorted = df.sort_values(by=(5))
df_sorted
```
