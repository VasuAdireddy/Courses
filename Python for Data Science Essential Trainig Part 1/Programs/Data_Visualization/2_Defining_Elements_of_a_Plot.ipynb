```python
# Steps to object oriented plotting
'''
1) Create a blank Figure object
2) Add Axes to the Figure
3) Generate plots within figure object
4) SPecify plotting and layout parameters for the plot within your figure
'''
```

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
#tells python to plot matplotlib plots inline in jupyter notebook
%matplotlib inline
rcParams['figure.figsize']=5,4  #defining figure and figuresize
```

### Defining axes,ticks and grids

```python
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]

fig=plt.figure() #creating figure
ax=fig.add_axes([.1,.1,1,1])  #adding axes to figure
ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])

#setting x and y limits
ax.set_xlim([1,9])
ax.set_ylim([1,3])

ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])

#setting x and y ticks
ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])

#adding grid
ax.grid()

ax.plot(x,y)
```

### Subplots

```python
fig=plt.figure()
fig, (ax1,ax2) = plt.subplots(1,2) #dividing figure into subplots

ax1.plot(x) #plot for x
ax2.plot(y) #plot for y
```
