```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize']=8,4
```

# Labeling Plot features

### Functional Method

```python
x=range(1,10)
y=[1,2,3,4,5,6,5,4,3]
plt.bar(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
z=[1,1.5,2,3,4.5,3.3]
labels=['cars','bikes','bicycles','buses','jeeps','trucks']
plt.pie(z,labels=labels)
plt.show()
```

### Object Oriented Method

```python
add=r"D:\Courses\New folder\Courses\Python for Data Science Essential Training Part 1\Ex_Files_Python_Data_Science_EssT_Pt_1\Ex_Files_Python_Data_Science_EssT_Pt_1\Exercise Files\Data\mtcars.csv"
cars=pd.read_csv(add)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg=cars.mpg

fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car_names')
ax.set_ylabel('miles/gal')
```

## Adding Legend

### Functional Method

```python
plt.pie(z)
plt.legend(labels,loc='best')
plt.show()
```

### Object oriented Method

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car_names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```
