```python
'''
method used to find correlation between linearly related continuous variables

Pearson Correlation Coeffecient
----used only for linear relationships
-> if R = 1 ---> strongly correlated
->    R = 0 ---> No linear relation
->    R = -1 --> strong negatice correlation (inverse)

'''
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

## Pearson Correlation

```python
add=r"D:\Courses\New folder\Courses\Python for Data Science Essential Training Part 1\Ex_Files_Python_Data_Science_EssT_Pt_1\Ex_Files_Python_Data_Science_EssT_Pt_1\Exercise Files\Data\mtcars.csv"
cars=pd.read_csv(add)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
sb.pairplot(cars)
```

```python
x= cars[['mpg','hp','qsec','wt']]
sb.pairplot(x)
```

## using scipy

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient,p_value = pearsonr(mpg,hp)
print("PearsonR coefficient correlation = %0.3f"%(pearsonr_coefficient))
```

```python
pearsonr_coefficient,p_value = pearsonr(mpg,qsec)
print("PearsonR coefficient correlation = %0.3f"%(pearsonr_coefficient))
```

```python
pearsonr_coefficient,p_value = pearsonr(mpg,wt)
print("PearsonR coefficient correlation = %0.3f"%(pearsonr_coefficient))
```
