```python
'''
used to find correlation between no linear, non normal and categorical variables

1) Spearmans rank correlation
--this has R values same as pearsonr correlation
--here assumes the variables are ordinal
2) Chi-square tables
-- p<0.05 -> reject null hypothesis and conclude variables are correlated
-- p>0.05 -> accept null hypothesis and conclude variables are independent
-- we need to do binning on numerical variables to convert them into categorical variables
'''
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14,7
plt.style.use('seaborn-whitegrid')
```

## Spearman Correlation

```python
add=r"D:\Courses\New folder\Courses\Python for Data Science Essential Training Part 1\Ex_Files_Python_Data_Science_EssT_Pt_1\Ex_Files_Python_Data_Science_EssT_Pt_1\Exercise Files\Data\mtcars.csv"
cars=pd.read_csv(add)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['cyl','vs','am','gear']]
sb.pairplot(x)
```

```python
## we can see all above are non linear, non normal
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient,p_values = spearmanr(cyl,vs)
print("Spearman Rank Correlation Coefficient %0.3f"%(spearmanr_coefficient))
```

```python
spearmanr_coefficient,p_values = spearmanr(cyl,am)
print("Spearman Rank Correlation Coefficient %0.3f"%(spearmanr_coefficient))
```

```python
spearmanr_coefficient,p_values = spearmanr(cyl,gear)
print("Spearman Rank Correlation Coefficient %0.3f"%(spearmanr_coefficient))
```

```python
# cly and vs have more correlation but negative
```

## Chi-Square test

```python
tabel = pd.crosstab(cyl,am)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabel.values)
print("Chi-Square statistics %0.3f.. p_values %0.3f"%(chi2,p))
```

```python
tabel = pd.crosstab(cyl,vs)

chi2, p, dof, expected = chi2_contingency(tabel.values)
print("Chi-Square statistics %0.3f.. p_values %0.3f"%(chi2,p))
```

```python
tabel = pd.crosstab(cyl,gear)

chi2, p, dof, expected = chi2_contingency(tabel.values)
print("Chi-Square statistics %0.3f.. p_values %0.3f"%(chi2,p))
```

```python
# above all p_values are less than 0.05, 
# so we conclude corresponding variables are correlated
```
