### In python missing values are treated as Nan

```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
```

```python
missing = np.nan
series_obj = Series(['row 1','row 2',missing,'row 4','row 5','row 6',missing,'row 8'])
series_obj
```

```python
series_obj.isnull()
```

```python
series_obj.notnull()
```

### Filling

```python
np.random.seed(25)
DF_obj =DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:5, 5] = missing
DF_obj
```

```python
filled_DF=DF_obj.fillna(0)
filled_DF
```

```python
filled_DF = DF_obj.fillna({0:0.1, 5:1.25})  #here we are filling specific column with specific values
filled_DF
```

```python
filled_DF= DF_obj.fillna(method = 'ffill')
filled_DF
# this will fill all NaN values in a cloumn with last non NaN value
```
