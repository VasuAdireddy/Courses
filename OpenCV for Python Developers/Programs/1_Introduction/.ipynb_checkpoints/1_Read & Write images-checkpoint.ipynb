```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
## reading image

img = cv2.imread("824696.jpg")
```

```python
## convering image to viewable rgb format from nparray matrices format

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
## we can also use google colab imshow function to show images
```

```python
## writing image to folder

cv2.imwrite("output.png",img)
```
