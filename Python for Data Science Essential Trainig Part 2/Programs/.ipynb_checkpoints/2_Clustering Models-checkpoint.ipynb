### K-Means method

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
iris = datasets.load_iris()

X = scale(iris.data)
y = pd.DataFrame(iris.target)  #predictant
variable_names = iris.feature_names
X[:10]
```

##### Building and running your model

```python
clustering = KMeans(n_clusters=3,random_state = 5)

clustering.fit(X)
```

##### Plotting model o/p

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
y.columns = ['Targets']
```

```python
color_theme = np.array(['darkgray','lightsalmon','powderblue'])

plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[iris.target], s=50)
plt.title('Ground Truth Classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[clustering.labels_], s=50)
plt.title('K-Means Classification')
```

```python
relabel = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)

plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[iris.target], s=50)
plt.title('Ground Truth Classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[relabel], s=50)
plt.title('K-Means Classification')
```

##### Evaluate clustering results

```python
print(classification_report(y,relabel))
```

```python
'''
precesion is a measure of models relevancy
recall is a measure of models completeness
high precision + high recall = highly accurate model results
'''
```

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
address = r"D:\Courses\New folder\Courses\Python for Data Science Essential Trainig Part 2\Ex_Files_Python_Data_Science_EssT_Pt2\Exercise Files\Data\mtcars.csv"

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

X = cars[['mpg', 'disp', 'hp', 'wt']].values

y = cars.iloc[:,(9)].values
```

##### using scipy to generate dendrogram

```python
z = linkage(X,'ward')
```

```python
dendrogram(z,truncate_mode='lastp',p=12,leaf_rotation=45,leaf_font_size = 15, show_contracted=True)

plt.title('Truncated Hierarchical CLustering Diagram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

# horizontal lines for determining cluster points
plt.axhline(y=500)
plt.axhline(y=150)
plt.show()
```

```python
# we choose 2 as cluster size
```

##### generate hierarchical cluster

```python
k = 2

Hclustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
Hclustering.fit(X)

sm.accuracy_score(y,Hclustering.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='average')
Hclustering.fit(X)

sm.accuracy_score(y,Hclustering.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='manhattan', linkage='average')
Hclustering.fit(X)

sm.accuracy_score(y,Hclustering.labels_)
```

```python
# like above, we can check all parameters and see the scores and select the best
```

### DBSCAN for outlier Detection

```python
import sklearn
from sklearn.cluster import DBSCAN
from collections import Counter
```

##### Train model and identify outliers

```python
add = r"D:\Courses\New folder\Courses\Python for Data Science Essential Trainig Part 2\Ex_Files_Python_Data_Science_EssT_Pt2\Exercise Files\Data\iris.data.csv"
df = pd.read_csv(add)
df.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']

data = df.iloc[:,0:4].values
target = df.iloc[:,4].values #species is our target

df[:5]
```

```python
# eps and min_samples are explained in theory file
model =DBSCAN(eps=0.8,min_samples=19).fit(data)
print(model)
```

```python
out_df = pd.DataFrame(data)
print(Counter(model.labels_))

# labels with -1 are outliers
print(out_df[model.labels_ == -1])
```

```python
# we only have 6 outliers, above we can see that
```

```python
fig=plt.figure()
ax = fig.add_axes([.1,.1,1,1])

colors = model.labels_
ax.scatter(data[:,2],data[:,1],c=colors,s=120)
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
plt.title('DBSCAN for Outlier Detection')
```

```python
# from the above visualization points with blue color are outliers, ie 6
```
