## Zeros()

```python
import numpy as np
```

```python
zarr = np.zeros(10)
```

```python
print(zarr) #we get float zeros because inbuilt fuctionality of zeros function gives only float
```

```python
zarr = zarr.astype(int)  #converts float 0 into int
```

```python
print(zarr)
```

```python
l = [1.2,2.3,3.4,4.5]
```

```python
l=np.array(l)
```

```python
print(l)
```

```python
l=l.astype(int)
```

```python
print(l)
```

## ones()

```python
oarr=np.ones(10)
```

```python
print(oarr)
```

```python
oarr=oarr.astype(int)
```

```python
print(oarr)
```

## full()

```python
farr=np.full(10,5)
```

```python
print(farr)
```

```python
print(farr)
```
