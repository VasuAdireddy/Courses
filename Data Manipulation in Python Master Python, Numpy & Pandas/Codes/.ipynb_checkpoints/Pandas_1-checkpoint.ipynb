#### Data frame is two dimensional
#### Series is one dimensional

```python
# In DataFrame
# each row is represented by "index"
# each column is represented by "column"
# row and column can be numbers or strings
# each column in a DataFrame is a Series
```

```python
# In Series
# it can have single row or column
# rows are represented by "index"
```

```python
import pandas as pd
import numpy as np
```

```python
mylist = [['Apple','Orange'],['Black','red']]
myDF = pd.DataFrame(mylist)
myDF
```

```python
# giving custom column labels
myDF = pd.DataFrame(mylist,columns=['Fruits','Colors'])
myDF
```

```python
#Creating DataFrame using np array
nplist = np.array([[1,2],[3,4],[5,6]])
myDF = pd.DataFrame(nplist)
myDF
```

```python
#Creating DataFrame using Dictionary
# each key will become column label
# each key should have one or more values
mydic = {'Fruit':['Apple','Orange','Banana'],'Color':['Red','Orange','Yellow']}
myDF = pd.DataFrame(mydic)
myDF
```
