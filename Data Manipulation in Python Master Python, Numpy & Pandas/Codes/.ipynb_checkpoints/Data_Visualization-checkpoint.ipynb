```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plots

```python
# takes 2 lists 1 for x axis and another for y axis

x = np.array([1,2,3])
y = np.array([4,5,6])

plt.plot(x,y)
```

```python
plt.plot(x,y, 'g')   #here g for green colour
```

```python
plt.title('Line Graph')
plt.plot(x,y, 'y')
```

```python
plt.title('Line Graph')
plt.xlabel('X_AXIS')
plt.ylabel('Y_AXIS')
plt.plot(x,y, 'b')
```

```python
plt.plot(x,y, 'v')  # its giving points graph
```

```python
# ploting 2 lines

x1 = np.array([10,52,35])
y1 = np.array([8,10,90])

x2 = np.array([10,25,100])
y2 = np.array([42,55,41])

plt.title('Two plots')
plt.plot(x1,y1, 'g')
plt.plot(x2,y2, 'y')
```

```python
plt.title('Two plots')
plt.plot(x1,y1, 'g')
plt.plot(x2,y2, 'y')

plt.legend(['plot1','plot2'])  #this will add a legend
```

## Plotting Histograms

```python
# A histogram is generally used to plot frequency which helps identify distribution of data

values = [10,15,20,10,15]
plt.hist(values)
```

```python
# in the above histogram, x axis is the values, y is the frequency of their occurances
```

```python
plt.hist(values, color = 'g', width = 0.5)
```

## Plotting Bar Charts

```python
# bar chart will have two arguments
# first is the x values on x axis
# on the yaxis we can give a list of values (one for one), which are the heights of the each graph
# we can also give a list of values for one x values
```

```python
val=[10,15,20, 12, 17]
plt.bar(val,[5,25,15,12,19], width = 0.5, color='g')
```

## Plotting Pie Charts

```python
values = [10,15,3,15,25]
plt.pie(values)       

#it will automatically calculate each value angle and plot accordingly
```

```python
plt.pie(values, labels = ['a','b','c','d','e']) 
```

```python
plt.pie(values, labels = ['a','b','c','d','e'],explode = [0,0,0.2,0,0]) 
```

```python
plt.pie(values, labels = ['a','b','c','d','e'], colors = ['r','b','y','g','pink']) 
```

## Plotting Log Plots

```python
# we can change the scale from linear to logarthmic
# if the values are more than other axis then this is used
```

```python
# plotting using logarthmic

x = np.array([1,2,3,4])
y = np.array([1,10,100,1000])
plt.yscale('log')
plt.plot(x,y)
```

```python
# plotting normally
# you can see the difference
plt.plot(x,y)
```

## Plotting Polar Plots

```python
# A point in a polar plot is represented as (r,theta)
# r is the distance from origin
# theta is the angle along which r is measured
```

```python
theta = np.arange(0,(2 * np.pi), 0.01)  # range (0- 2*3.14) step-0.01
r = 2
for i in theta:
    plt.polar(i,r)
```

## Handling Dates


```python
# Sometimes there will be lots of values in x axis, it will become dificult to plot all, and they might overlap
# in such cases we can rotate the orientation on x values , for better visualization
```

```python
dates = np.array(['2018-01-01','2019-01-01','2020-01-01','2021-01-01','2022-01-01',
                 '2023-01-01','2024-01-01','2025-01-01','2026-01-01','2027-01-01',])
values = np.array([100,101,102,103,105,106,107,108,109,110])

plt.xticks(rotation = 90)
plt.plot(dates,values)
```

```python
# above we can see the dates orientation changes
```

## Creating Multiple Subplots in one figure


```python
# .subplot() before ploting will do this job
# it has 3 arguments -- 1: no of rows 2: no of columns 3: position of plot
```

```python
x1 = [2,4,6,8]
y1 = [1,10,100,1000]

x2=[1,3,5,7]
y2=[100,110,120,130]

plt.subplot(1,2,1)
plt.plot(x1,y1,'r')

plt.subplot(1,2,2)
plt.plot(x2,y2,'y')

plt.tight_layout() #used to adjust the space between 2 plots

# we can also have seperate titles, xlabels and ylabels
```
